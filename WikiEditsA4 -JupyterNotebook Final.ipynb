Abstract:
A wikipedia edit is made up of three components: a base sentence, a phrase, and a edited sentence. For this datathon, we decided to focus on the insertion data of Wikiedits.

**Research Question:** Given a base sentence and a phrase, is it possible to accurately identify where the phrase should go?

We used Pandas to manipulate data, nltk to classify parts of speech, and sklearn for machine learning. In the end, we were able to achieve 28% accuracy on 2,000 cases unseen by our model after feeding it 18,000 cases.

```python
def subSet(fullSet, subSet):
    for i in range(len(fullSet) - len(subSet) + 1):
        if (fullSet[i] == subSet[0]) :
            isValid = True;
            for j in range (len(subSet)):
                if (fullSet[i+j] != subSet[j]):
                    isValid = False
                    break
            if (isValid):
                return i;
    return -1;
```

## Extracting Data

From each of the cases, we extracted data that would be used as a feature vector for our ML model. The following is a list of the features.

1. Base sentence word count
2. Phrase word Count
3. Phrase Index - This is the OUTCOME of the feature vector
4. Part of speech of word 2 position before phrase starts
5. Part of speech of word 1 position before phrase starts
6. Part of speech of first word in phrase
7. Part of speech of second word in phrase
8. Part of speech of second to last word in phrase
9. Part of speech of last word in phrase
10. Part of speech of word 1 position after phrase ends
11. Part of speech of word 2 positions after phrase ends
12. Number of nouns in phrase
13. Number of commas in phrase
14. Number of nouns in base sentence
15. Number of commas in base sentence
16. Capitalization of first word in base sentence
17. Capitalization of word 2 position before phrase starts
18. Capitalization of word 1 position before phrase starts
19. Capitalization of first word in phrase
20. Capitalization of second word in phrase
21. Capitalization of second to last word in phrase
22. Capitalization of last word in phrase
23. Capitalization of word 1 position after phrase ends
24. Capitalization of word 2 positions after phrase ends
25. Part of speech of first word in base sentence
26. Part of speech of last word in base sentence
27. Number of nouns before phrase
28. Number of commas before phrase
29. Number of nouns after phrase
30. Number of commas after phrase

```python
insertions1 = insertions

#3 - Base Statement - Word Count
insertions1['bs_word_count'] = 0
#4 - Phrase - Word Count
insertions1['p_word_count'] = 0
#5 - Phrase - Index in Edited Statement (THIS IS THE WHAT WE PREDICT)
insertions1['p_index'] = -1

#6 - PoS Before before Phrase
insertions1['p-2_pos'] = ""
#7 - PoS Before Phrase
insertions1['p-1_pos'] = ""
#8 - PoS Beginning of Phrase
insertions1['p_begin_pos'] = ""
#9 - PoS After Beginning of Phrase
insertions1['p_begin_after_pos'] = ""
#10 - PoS Before End of Phrase
insertions1['p_end_before_pos'] = ""
#11 - PoS End of Phrase
insertions1['p_end_pos'] = ""
#12 - PoS After Phrase
insertions1['p+1_pos'] = ""
#13 - PoS After After Phrase
insertions1['p+2_pos'] = ""


#14 - Phrase - Number of Nouns
insertions1['num_nouns_p'] = 0
#15 - Phrase - Number of Commas
insertions1['num_commas_p'] = 0
#16 - Base Sentence - Number of Nouns
insertions1['num_nouns_bs'] = 0
#17 - Base Sentence - Number of Commas
insertions1['num_commas_bs'] = 0

#18
insertions1['bs_first_capital'] = False
#19
insertions1['p-2_capital'] = False
#20
insertions1['p-1_capital'] = False
#21
insertions1['p_begin_capital'] = False
#22
insertions1['p_begin_next_capital'] = False
#23
insertions1['p_end_before_capital'] = False
#24
insertions1['p_end_capital'] = False
#25
insertions1['p+1_capital'] = False
#26
insertions1['p+2_capital'] = False

#27
insertions1['begin_pos'] = ""
#28
insertions1['end_pos'] = ""


#29
insertions1['num_nouns_before'] = 0
#30
insertions1['num_commas_before'] = 0
#31
insertions1['num_nouns_after'] = 0
#32
insertions1['num_commas_after'] = 0

for x in range(len(insertions1)):
    #Percent Tracker
    if(x%100 == 0):
        print(str(x/100)+"%")
        
    #Adds Base Sentence Information
    baseWord = insertions.iloc[x,0].split()
    insertions.iloc[x,3] = len(baseWord)
    for y in range(len(baseWord)):
        #Adds #16 and #17 (Base Sentence Number of Nouns and Commas)
        if(nltk.pos_tag(nltk.word_tokenize(str(baseWord[y])))[0][1][0] == 'N'):
            insertions1.iloc[x,16] = insertions1.iloc[x,16] + 1
        elif(nltk.pos_tag(nltk.word_tokenize(str(baseWord[y])))[0][1][0] == ","):
            insertions1.iloc[x,17] = insertions1.iloc[x,17] + 1
    
    #Turns Edited Statement and Phrase into Lists of strings
    editedWord = insertions1.iloc[x,2].split()
    phraseWord = insertions1.iloc[x,1].split()
    
    
    #Adds Phrase Information
    #Adds #4 (Phrase Length)
    insertions.iloc[x,4] = len(phraseWord)
    #Adds #5 (Phrase Index)
    insertions.iloc[x,5] = subSet(editedWord, phraseWord)
    
    firstWords = []
    lastWords = []
    for y in range(len(editedWord)):
        if (y < insertions.iloc[x,5]):
            firstWords.append(editedWord[y])
        elif (y > insertions.iloc[x,5]):
            lastWords.append(editedWord[y])
    
    
    for y in range(len(firstWords)):
        #Adds #29 and #30 (The part before the phrase of Nouns and Commas)
        if(nltk.pos_tag(nltk.word_tokenize(str(firstWords[y])))[0][1][0] == 'N'):
            insertions1.iloc[x,29] = insertions1.iloc[x,29] + 1
        elif(nltk.pos_tag(nltk.word_tokenize(str(firstWords[y])))[0][1][0] == ","):
            insertions1.iloc[x,30] = insertions1.iloc[x,30] + 1
    
    for y in range(len(lastWords)):
        #Adds #31 and #32 (The part after the phrase Number of Nouns and Commas)
        if(nltk.pos_tag(nltk.word_tokenize(str(lastWords[y])))[0][1][0] == 'N'):
            insertions1.iloc[x,31] = insertions1.iloc[x,31] + 1
        elif(nltk.pos_tag(nltk.word_tokenize(str(lastWords[y])))[0][1][0] == ","):
            insertions1.iloc[x,32] = insertions1.iloc[x,32] + 1
    
    #Adds #6 (PoS Before Phrase)
    if(insertions.iloc[x,5] > 1):
        insertions.iloc[x,6] = nltk.pos_tag(nltk.word_tokenize(str(editedWord[insertions.iloc[x,5]-2])))[0][1]
        insertions1.iloc[x,19] = editedWord[insertions.iloc[x,5]-2][0].isupper()
    #Adds #7 (PoS Before Phrase)
    if( insertions.iloc[x,5] > 0):
        insertions.iloc[x,7] = nltk.pos_tag(nltk.word_tokenize(str(editedWord[insertions.iloc[x,5]-1])))[0][1]
        insertions.iloc[x,20] = editedWord[insertions.iloc[x,5]-1][0].isupper()
    #Adds #8 (PoS Beginning of Phrase)
    insertions1.iloc[x,8] = nltk.pos_tag(nltk.word_tokenize(str(phraseWord[0])))[0][1]
    insertions.iloc[x,21] = phraseWord[0][0].isupper()
    #Adds #11 (PoS End of Phrase)
    insertions1.iloc[x,11] = nltk.pos_tag(nltk.word_tokenize(str(phraseWord[len(phraseWord)-1])))[0][1]
    insertions.iloc[x,24] = phraseWord[len(phraseWord)-1][0].isupper()
    #Adds #12 (PoS End of Phrase)
    if(insertions.iloc[x,5] + len(phraseWord) < len(editedWord) - 1 ):
        insertions.iloc[x,12] = nltk.pos_tag(nltk.word_tokenize(str(editedWord[insertions.iloc[x,5] + len(phraseWord)+1])))[0][1]
        insertions.iloc[x,25] = editedWord[insertions.iloc[x,5] + len(phraseWord)+1][0].isupper()
    if(insertions.iloc[x,5] + len(phraseWord) < len(editedWord) - 2 ):
        insertions.iloc[x,13] = nltk.pos_tag(nltk.word_tokenize(str(editedWord[insertions.iloc[x,5] + len(phraseWord)+2])))[0][1]
        insertions.iloc[x,26] = editedWord[insertions.iloc[x,5] + len(phraseWord)+2][0].isupper()
    if(len(phraseWord) >= 2):
        #Adds #8 (PoS After Beginning of Phrase)
        insertions1.iloc[x,9] = nltk.pos_tag(nltk.word_tokenize(str(phraseWord[1])))[0][1]
        insertions.iloc[x,22] = phraseWord[1][0].isupper()
        #Adds #9 (PoS Before End of Phrase)
        insertions1.iloc[x,10] = nltk.pos_tag(nltk.word_tokenize(str(phraseWord[len(phraseWord)-2])))[0][1]
        insertions.iloc[x,23] = phraseWord[len(phraseWord)-2][0].isupper()
    for y in range(len(phraseWord)):
        #Adds #12 and #13 (Phrase - Number of Nouns and Commas)
        if(nltk.pos_tag(nltk.word_tokenize(str(phraseWord[y])))[0][1][0] == 'N'):
            insertions1.iloc[x,14] = insertions1.iloc[x,14] + 1
        elif(nltk.pos_tag(nltk.word_tokenize(str(phraseWord[y])))[0][1][0] == ","):
            insertions1.iloc[x,15] = insertions1.iloc[x,15] + 1

        
    insertions1.iloc[x,18] = baseWord[0][0].isupper()
    insertions.iloc[x,27] = nltk.pos_tag(nltk.word_tokenize(str(editedWord[0])))[0][1]
    insertions.iloc[x,28] = nltk.pos_tag(nltk.word_tokenize(str(editedWord[len(editedWord)-2])))[0][1]
    
```

## Creating Feature Matrix

```python
tags = ['CC','CD','DT','EX','FW','IN','JJ','JJR','JJS','LS','MD','NN','NNS','NNP','NNPS','PDT','POS','PRP','PRP$','RB','RBR','RBS','RP','TO','UH','VB','VBD','VBG','VBN','VBP','VBZ','WDT','WP','WP$','WRB']
X2 = np.empty([len(insertions1),30])
Y2 = np.empty([len(insertions1),])

#Creating Feature Vector
for x in range(len(insertions)):
    #Base Sentence Length
    X2[x,0] = int(insertions1.iloc[x,3])
    #Phrase Length
    X2[x,1] = int(insertions1.iloc[x,4])
    #PoS Before Phrase
    for y in range(8):
        if(insertions1.iloc[x,6+y] in tags):
            X2[x,2+y] = tags.index(insertions1.iloc[x,6+y])
        else:
            X2[x,2+y] = 35
    
    for y in range(13):
        X2[x,10+y] = int(insertions1.iloc[x,14+y])
    
    if(insertions1.iloc[x,28] in tags):
            X2[x,24] = tags.index(insertions1.iloc[x,28])
    else:
        X2[x,24] = 35
    
    if(insertions1.iloc[x,29] in tags):
        X2[x,25] = tags.index(insertions1.iloc[x,29])
    else:
        X2[x,25] = 35
    
    for y in range(4):
        X2[x,26+y] = int(insertions1.iloc[x,29+y])
        
    Y2[x] = int(insertions1.iloc[x,5])
```
