```python
insertions1 = insertions
insertions1['bs_word_count'] = ""
insertions1['p_word_count'] = insertions1['phrase'].str.count(' ') + 1
insertions1['es_word_count'] = ""
insertions1['p_punctuation'] = insertions1['phrase'].str.contains("""[!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~1234567890]""") == False
insertions1['p_index'] = -1
insertions1['p_pos'] = ""
insertions1['p-1_pos'] = ""
insertions1['p+1_pos'] = ""
insertions1['num_nouns'] = 0
insertions1['num_commas'] = 0

for x in range(len(insertions1)):
    if(x%1000):
        print(str(x/1000)+"%")
    baseWord = insertions.iloc[x,0].split()
    insertions.iloc[x,3] = len(baseWord)
    for y in range(len(baseWord)):
        if(nltk.pos_tag(nltk.word_tokenize(str(baseWord[y])))[0][1][0] == 'N'):
            insertions1.iloc[x,11] = insertions1.iloc[x,11] + 1
        elif(nltk.pos_tag(nltk.word_tokenize(str(baseWord[y])))[0][1][0] == ","):
            insertions1.iloc[x,12] = insertions1.iloc[x,12] + 1
    editedWord = insertions.iloc[x,2].split()
    insertions.iloc[x,5] = len(editedWord)
    if(insertions1.iloc[x,4]==1):
        insertions1.iloc[x,7] = editedWord.index(insertions1.iloc[x,1])
        insertions1.iloc[x,8] = nltk.pos_tag(nltk.word_tokenize(str(insertions1.iloc[x,1])))[0][1]
        if(not insertions.iloc[x,7] == 0):
            insertions1.iloc[x,9] = nltk.pos_tag(nltk.word_tokenize(str(editedWord[insertions1.iloc[x,7]-1])))[0][1]
        if(insertions1.iloc[x,7] <= insertions.iloc[x,5] - 1):
            insertions1.iloc[x,10] = nltk.pos_tag(nltk.word_tokenize(str(editedWord[insertions1.iloc[x,7]+1])))[0][1]
```
